```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
populationdist = pd.read_csv("unkown분포.csv")
unkowndist = pd.read_csv("전체분포.csv")
```

```python
populationdist.info()
```

```python
unkowndist.info()
```

```python
# 전체 금액 분포
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 원본 분포
populationdist["amount"].hist(bins=30, ax=axes[0])
axes[0].set_title("Population Original Distribution")
axes[0].set_xlabel("amount")
axes[0].set_ylabel("Frequency")
axes[0].grid(False)

# 로그 변환 분포
log_amount = np.log1p(populationdist["amount"])
axes[1].hist(log_amount, bins=30)
axes[1].set_title("Log Distribution")
axes[1].set_xlabel("log(amount)")
axes[1].set_ylabel("Frequency")
axes[1].grid(False)

plt.tight_layout()
plt.show()
```

```python
# Unknown 금액 분포
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 원본 분포
unkowndist["amount"].hist(bins=30, ax=axes[0])
axes[0].set_title("Unknown - Original Distribution")
axes[0].set_xlabel("amount")
axes[0].set_ylabel("Frequency")
axes[0].grid(False)

# 로그 분포
log_amount = np.log1p(unkowndist["amount"])
axes[1].hist(log_amount, bins=30)
axes[1].set_title("Unknown - Log Distribution")
axes[1].set_xlabel("log(amount)")
axes[1].set_ylabel("Frequency")
axes[1].grid(False)

plt.tight_layout()
plt.show()
```

```python
# 전체 reward 분포

populationdist["Reward"].hist(bins=30)
plt.title("Original Distribution")
plt.xlabel("Reward")
plt.ylabel("Frequency")
plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
# Unknown reward 분포

unkowndist["Reward"].hist(bins=30)
plt.title("Unknown - Distribution")
plt.xlabel("Reward")
plt.ylabel("Frequency")
plt.grid(False)


plt.tight_layout()
plt.show()
```

```python
# 전체 event 분포

populationdist["event"].value_counts().plot(kind="bar")
plt.title("Original Distribution")
plt.xlabel("event")
plt.ylabel("Frequency")
plt.grid(False)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Unknown event 분포

unkowndist["event"].value_counts().plot(kind="bar")
plt.title("Unknown - Distribution")
plt.xlabel("event")
plt.ylabel("Frequency")
plt.grid(False)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 전체 time 분포

populationdist["time"].hist(bins=30)
plt.title("Original Distribution")
plt.xlabel("time")
plt.ylabel("Frequency")
plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
# Unknown time 분포

unkowndist["time"].hist(bins=30)
plt.title("Unknown - Distribution")
plt.xlabel("time")
plt.ylabel("Frequency")
plt.grid(False)


plt.tight_layout()
plt.show()
```
