```python
merge_df = pd.merge(
    transcript_df,
    portfolio_df,
    left_on=transcript_df['Record Value'],
    right_on=portfolio_df['promo_id'],
    how='left'
)

if 'key_0' in merge_df.columns :
    merge_df.drop(columns=['key_0'], inplace=True)
merge_df
```

```python
offer_df = merge_df[merge_df['Record'] == 'offer_id'].copy()
offer_df = offer_df.sort_values(['customer_id', 'promo_id', 'time'])
```

```python
offer_df['sequence'] = offer_df.groupby(['customer_id', 'promo_id'])['time'].rank(method='first')
```

```python
sequence = offer_df.pivot_table(
    index=['customer_id', 'promo_id'],
    columns='event',
    values='time',
    aggfunc='min'
).reset_index()

column_order = ['customer_id', 'promo_id', 'offer received', 'offer viewed', 'offer completed']

existing_cols = [col for col in column_order if col in sequence.columns]
sequence = sequence.reindex(columns=existing_cols)
sequence
```
