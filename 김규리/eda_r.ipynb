```python
# 각 단계별 총 인원 합계
n_reg = df['registered'].sum()
n_view = df['viewed'].sum()
n_exp = df['explored'].sum()

# 등록자 중 실제 강의를 본 비율
reg_to_view = (n_view / n_reg) * 100
# 이탈률
drop_out_reg_to_view = 100 - reg_to_view

# 강의를 본 사람 중 탐색한 비율
view_to_exp = (n_exp / n_view) * 100
# 이탈률
drop_out_view_to_exp = 100 - view_to_exp

# 전체 등록자 대비 탐색 비율
total_conversion = (n_exp / n_reg) * 100

print(f"========== 단계별 참여 및 이탈 구조 ==========")
print(f"registered -> viewed 전환율 : {reg_to_view:.2f}%")
print(f"registered -> viewed 이탈률 : {drop_out_reg_to_view:.2f}%")
print()
print(f"viewed -> explored 전환율 : {view_to_exp:.2f}%")
print(f"viewed -> explored 이탈률 : {drop_out_view_to_exp:.2f}%")
print()
print(f"최종 explored 전환율 : {total_conversion:.2f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = df.reset_index(drop=True)

# 임계값 설정
thresholds = {
    'ndays_act': 18,
    'nevents': 1274,
    'nplay_video': 170
}

# step 2 유저 그룹 분류 
# viewed는 했지만 explored는 못한 유저들 중, 설정한 기준을 하나라도 넘긴 '고관여 탐색 이탈자' 추출
target_condition = (df['viewed'] == 1) & (df['explored'] == 0)
df['is_high_effort_leaver'] = target_condition & ( 
    (df['ndays_act'] >= thresholds['ndays_act']) |
    (df['nevents'] >= thresholds['nevents']) |
    (df['nplay_video'] >= thresholds['nplay_video'])
)

# 임계값 기준 분포 확인
features = ['ndays_act', 'nevents', 'nplay_video']
plt.figure(figsize=(15, 10))

for i, col in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='is_high_effort_leaver', y=col, data=df[target_condition])
    plt.axhline(thresholds[col], color='red', linestyle='--', label=f'Threshold ({thresholds[col]})')
    plt.title(f'Distribution of {col} (Viewed but not Explored)')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 이탈자 유형 세분화
# Step 1. 진입 전 이탈
r_step1 = ((df['registered'] == 1) & (df['viewed'] == 0)).sum()

# Step 2. 탐색 중 이탈
# 2-1. 단순 탐색 후 이탈 (기준치 미달)
r_step2_1 = ((df['viewed'] == 1) & (df['explored'] == 0) & (df['is_high_effort_leaver'] == False)).sum()

# 2-2. 고관여 탐색 후 이탈 (기준치 초과)
r_step2_2 = ((df['viewed'] == 1) & (df['explored'] == 0) & (df['is_high_effort_leaver'] == True)).sum()

print(f"진입 전 이탈 : {r_step1:,}명")
print(f"단순 탐색 후 이탈 : {r_step2_1:,}명")
print(f"고관여 탐색 후 이탈 : {r_step2_2:,}명")
```

```python
# Viewed -> Explored 단계 이탈자 중 고관여 탐색 유저의 비율
leaver_count = df[target_condition].shape[0]
high_effort_count = df['is_high_effort_leaver'].sum()
ratio = (high_effort_count / leaver_count) * 100

print(f"탐색 단계 이탈자 중 고관여 탐색 유저 비율: {ratio:.2f}%")
```

```python
# 이탈자 유형별 특성 분석을 위한 데이터프레임 생성
target_leaver_df = df[(df['viewed'] == 1) & (df['explored'] == 0)].copy()
high_effort_leaver = target_leaver_df[target_leaver_df['is_high_effort_leaver'] == True]
low_effort_leaver = target_leaver_df[target_leaver_df['is_high_effort_leaver'] == False]
```

```python
def compare_leaver_personas(feature, title, order=None):
    # 각 이탈자 유형별 비율 계산
    high_dist = high_effort_leaver[feature].value_counts(normalize=True).reindex(order).fillna(0)
    low_dist = low_effort_leaver[feature].value_counts(normalize=True).reindex(order).fillna(0)
    
    # 데이터프레임 재구성
    comp_df = pd.DataFrame({
        'Category': high_dist.index,
        'High-Effort': high_dist.values,
        'Low-Effort': low_dist.values
    }).melt(id_vars='Category', var_name='Group', value_name='Ratio')

    # 시각화
    plt.figure(figsize=(12, 6))
    sns.barplot(data=comp_df, x='Category', y='Ratio', hue='Group', palette=['#B20610', '#00262E'])
    plt.title(f'Persona Comparison: {title}', fontsize=15, pad=20)
    plt.ylabel('Proportion of Group', fontsize=12)
    plt.xlabel(feature, fontsize=12)
    plt.xticks(rotation=30)
    plt.legend(title='User Group')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
# 학력 비교 
loe_order = ['Less than Secondary', 'Secondary', "Bachelor's", "Master's", 'Doctorate', 'unknown']
compare_leaver_personas('LoE_DI', 'Level of Education', order=loe_order)

# 연령대 비교
age_order = ['<20', '20s', '30s', '40s', '50s', '60s+', 'unknown']
compare_leaver_personas('age_group', 'Age Group', order=age_order)

# 주요 수치 요약 출력
# 평균값과 함께 중앙값을 출력하여 이상치에 의한 왜곡이 없는지 대조
# 두 그룹 간의 활동량 차이가 상식적인 수준(임계값 설정 논리)과 부합하는지 최종 확인
print("\n" + "="*50)
print("고활동 vs 저활동 이탈자 활동량 비교 ")
print("="*50)
metrics = ['nplay_video', 'ndays_act', 'nevents']
summary = pd.DataFrame({
    'High_Effort_Avg': high_effort_leaver[metrics].mean(),
    'Low_Effort_Avg': low_effort_leaver[metrics].mean(),
    'Difference(X)': high_effort_leaver[metrics].mean() / low_effort_leaver[metrics].mean()
})
print(summary.round(2))
```

```python
from scipy.stats import chi2_contingency
# 교차표 생성
loe_contingency = pd.crosstab(target_leaver_df['is_high_effort_leaver'], target_leaver_df['LoE_DI'])

# 카이제곱 검정 실행
chi2, p_loe, dof, expected = chi2_contingency(loe_contingency)

print(f"\n" + "="*50)
print(f" [학력 분석 결과] 범주형 데이터 (LoE_DI) 기반 카이제곱 검정")
print(f" - p-value: {p_loe:.4f}")
print("-" * 50)

if p_loe < 0.05:
    print(" >> 결과: 두 집단의 학력 분포는 통계적으로 유의미한 차이가 있음 (학력이 이탈에 영향을 줌)")
    print(" >> 인사이트: 특정 학력 계층이 고활동 이탈을 더 많이 하는 경향이 있음 (타겟 분석 필요)")
else:
    print(" >> 결과:두 집단의 학력 분포 차이는 유의미하지 않음 (학력과 관계없이 공통적으로 발생)")
print("="*50)
```

```python
# 학력 효과 크기: Cramer's V
import numpy as np
def calculate_cramers_v(contingency_table):
    chi2 = chi2_contingency(contingency_table)[0]
    n = contingency_table.sum().sum()
    phi2 = chi2 / n
    r, k = contingency_table.shape
    return np.sqrt(phi2 / min(k-1, r-1))

print("\n" + "="*50)
print(" [효과 크기 분석 결과] 실질적 유의성 검증 ")
print("="*50)

loe_ct = pd.crosstab(target_leaver_df['is_high_effort_leaver'], target_leaver_df['LoE_DI'])
v = calculate_cramers_v(loe_ct)
print(f"- 학력(LoE_DI) Cramer's V: {v:.4f}")
```

```python
from scipy.stats import mannwhitneyu
stat_age, p_age = mannwhitneyu(high_effort_leaver['age'].dropna(), 
                               low_effort_leaver['age'].dropna(), 
                               alternative='two-sided')
print(f"\n" + "="*50)
print(f" [연령 분석 결과] 수치 데이터(age) 기반 비모수 검정")
print(f" - Mann-Whitney U p-value: {p_age:.4f}")
print("-" * 50)
if p_age < 0.05:
    print(" >> 결과: 두 집단의 연령 분포는 통계적으로 유의미한 차이가 있음 (나이가 이탈에 영향을 줌)")
else:
    print(" >> 결과: 두 집단의 연령 분포 차이는 유의미하지 않음 (나이와 관계없이 공통적으로 발생)")
print("="*50)
```

- 고활동 이탈은 유저의 인구통계학적 특성과 무관하게 발생함
- 이는 이탈의 원인이 유저 내부가 아닌 특정 강의의 난이도, 학습 구조, 콘텐츠 불친절함 등 외부 요인에 강력하게 쏠려 있음을 시사함
