### explored, certified 컬럼 정합성 검증 및 전처리

#### 개념 정의
- `explored`는 강의 코스웨어의 50% 이상에 접근한 학습자 여부 (0/1)
- 퍼널 상 `viewed` 이후 단계에 해당
→ 즉, `explored = 1`이면 반드시 `viewed = 1`이 선행되어야 함

#### 전처리 기준 (퍼널 정합성)
- 정상 흐름: `viewed → explored`
- 이 순서를 위반하는 데이터는 논리적으로 불가능한 케이스로 판단

#### 제거 대상
- `viewed = 0` & `explored = 1` → 8건
- 한 번도 열람하지 않고 50% 이상 학습하는 것은 불가능
→ 데이터 오류로 판단하여 제거

- `certified`는 행동 단계가 아닌 결과(성과) 변수로 판단
→ 퍼널에서 제외하고 보조 지표로 활용
- `explored = 0` & `certified = 1` → 약 690건
- 유지 결정

- `certified`가 퍼널에서 제외되었기 때문에 퍼널 분석에는 영향 없음
- 수료증만을 목표로 한 학습자 집단으로 해석 가능
→ 별도 분석 가치 존재

#### 최종 결론
- 퍼널 위반(논리적 오류) 데이터만 제거
- 행동 정의와 충돌하지 않는 데이터는 유지

```python
# 제거 대상만 따로 보기
invalid_funnel = course[
    ((course["explored"] == 1) & (course["viewed"] != 1))
]

len(invalid_funnel)
```

```python
course = course[
    ~(
        ((course["explored"] == 1) & (course["viewed"] != 1))
    )
]
```

### incomplete_flag 데이터 제거

#### 문제 정의
- incomplete_flag = 1인 데이터는 특정 변수에서 결측이 집중적으로 발생
- 특히 n_events는 대부분 결측이며, 다른 변수 또한 분포 차이를 보임

#### 결측 메커니즘 관점 해석
- 결측이 특정 그룹(incomplete_flag)에 집중되어 나타남
→ 이는 Missing Completely at Random(MCAR)이 아닌,
   **Missing at Random(MAR) 또는 Missing Not at Random(MNAR)** 구조로 판단됨

- 이러한 경우, 결측 자체가 데이터의 특성과 관련될 가능성이 높으며,
  분석 결과에 편향(bias)을 유발할 수 있음

#### 샘플 대표성 문제 (Selection Bias)
- incomplete_flag = 1 데이터를 포함할 경우,
  일부 사용자 그룹의 행동이 과소 또는 왜곡 반영될 가능성이 있음

- 특히 행동 로그 기반 분석에서는,
  관측 가능한 집단과 관측 불가능 집단이 혼재될 경우
  전체 사용자 행동을 대표하지 못하는 문제가 발생함

#### 식별성(Identifiability) 문제
- 행동 로그가 일부만 존재하는 경우,
  해당 사용자의 실제 행동 상태를 정의할 수 없음

- 이는 퍼널 단계 정의 및 KPI 산출 과정에서
  해석 불가능한 상태를 초래함

#### 처리 결정
→ 결측이 구조적으로 발생하며 분석 편향을 유발할 가능성이 높다고 판단하여,
   데이터의 일관성과 해석 가능성을 확보하기 위해
   `incomplete_flag = 1` 데이터를 분석에서 제외
- 단순 결측 대체(imputation)는 고려했으나,
  n_events와 같이 구조적으로 결측이 발생한 변수의 경우
  값 보정이 실제 행동을 반영하지 못할 가능성이 높다고 판단

- 또한 incomplete_flag=1 데이터는 일부 변수만 관측된 상태로,
  사용자 행동을 완전하게 정의할 수 없기 때문에
  부분적 보정보다는 제거가 더 적절하다고 판단

#### 결론
- 본 제거는 단순 결측 처리의 문제가 아니라,
  **데이터 생성 구조와 결측 메커니즘을 고려한 분석 설계 결정**임

```python
import pandas as pd
import numpy as np

# =========================
# 0. 활동 지표 컬럼 정의
# =========================
activity_cols = [
    "last_event_DI",
    "nevents",
    "ndays_act",
    "nplay_video",
    "nchapters",
    "nforum_posts"
]

# =========================
# 1. incomplete_flag 기본 분포 확인
# =========================
print("=== incomplete_flag 분포 ===")
print(course["incomplete_flag"].value_counts(dropna=False))
print()

print("=== incomplete_flag 비율(%) ===")
print((course["incomplete_flag"].value_counts(dropna=False, normalize=True) * 100).round(2))
print()

# =========================
# 2. incomplete_flag별 활동지표 결측 개수 / 결측 비율 확인
# =========================
print("=== incomplete_flag별 활동지표 결측 개수 ===")
missing_count = course.groupby("incomplete_flag")[activity_cols].apply(lambda x: x.isna().sum())
print(missing_count)
print()

print("=== incomplete_flag별 활동지표 결측 비율 ===")
missing_ratio = course.groupby("incomplete_flag")[activity_cols].apply(lambda x: x.isna().mean().round(4))
print(missing_ratio)
print()

```

### last_event_DI 기반 전처리

본 단계에서는 마지막 활동 시점(`last_event_DI`)을 기준으로
사용자의 활동 상태를 재정의하고,
시간 정합성을 만족하지 않는 데이터를 정제하였다.

---

#### 3.1 Registered-only 사용자 처리

- 일부 데이터는 등록 시점(`start_time_DI`) 이후
  추가적인 활동이 기록되지 않은 상태로 존재함

- `last_event_DI`가 결측이며 `viewed != 1`인 경우,
  등록 이후 별도의 학습 활동이 없었던 사용자로 해석

처리 방식:
- 활동 지표(`nevents`, `ndays_act`, `nplay_video`, `nchapters`, `nforum_posts`)를 0으로 설정
- `last_event_DI = start_time_DI`로 보정

→ 등록 후 즉시 이탈한 사용자로 정의

---

#### 3.2 Viewed-only 사용자 처리

- 일부 데이터는 `viewed = 1`로 기록되어 있으나,
  `last_event_DI` 및 활동 지표가 결측으로 존재

- 이는 최소한 강의에 접근한 이력은 있으나,
  구체적인 활동 로그가 관측되지 않은 경우로 해석됨

- 해당 데이터는 행동 발생 여부(`viewed`)는 명확하지만,
  활동 강도는 확인할 수 없는 **부분 관측 데이터**에 해당함

처리 방식:
- 활동 지표는 임의로 보정하지 않고 결측 상태를 유지
- 시간 정합성을 위해 `last_event_DI = start_time_DI`로 보정
- 필요 시 별도 플래그로 관리

→ 퍼널 상 "진입 후 이탈 사용자"로 해석하여 유지

#### 3.3 duration < 0 데이터 제거

- `last_event_DI < start_time_DI`인 경우 존재
- 이는 시간 순서가 역전된 비정상 데이터

처리 방식:
- `duration < 0`인 데이터 제거

→ 시간 정합성을 만족하는 데이터만 유지

---

#### 3.4 duration = 0 패턴 재검토

- `duration = 0`은 시작일과 마지막 활동일이 같은 날임을 의미하며,
  반드시 비정상적인 학습을 의미하지는 않음

- 온라인 강의 특성상, 당일에 집중적으로 학습하여
  `explored = 1` 또는 `certified = 1`에 도달하는 사례가 발생할 수 있음

- 초기에는 이를 비정상 데이터로 간주하여 제거를 고려하였으나,
  명확한 오류로 단정하기 어려워 제거하지 않기로 결정

처리 방식:
- 해당 데이터는 제거하지 않고 유지
- 필요 시 단기 완료 패턴으로 별도 분석 대상으로 활용

---

#### 추가 검토: duration과 ndays_act 관계

- `duration`은 전체 학습 기간(날짜 차이)을 의미하고,
  `ndays_act`는 실제 활동이 발생한 날짜 수를 의미함

- 두 변수는 서로 다른 개념(기간 vs 활동 빈도)을 측정하므로,
  직접 비교를 통해 이상치를 정의하는 것은 적절하지 않음

- 특히 동일 날짜 내 활동은 `duration = 0`으로 계산될 수 있어
  `ndays_act > duration`과 같은 패턴이 발생할 수 있음

→ 해당 관계를 기준으로 데이터 제거는 수행하지 않음

---

#### 결론

→ `last_event_DI`를 기준으로 사용자 행동을 재정의하고,
   시간 순서 정합성은 엄격히 검증하되,
   학습 패턴 다양성은 보존하는 방향으로 전처리를 수행함

```python
# 3.1, 3.2 코드 (수정본)

# 복사본
course_fixed = course_clean.copy()

# 조건 정의
mask_last_missing = course_fixed["last_event_DI"].isna()
mask_viewed_missing = mask_last_missing & (course_fixed["viewed"] == 1)
mask_non_viewed_missing = mask_last_missing & (course_fixed["viewed"] != 1)

# 활동성 관련 컬럼 지정
activity_cols = ["nevents", "ndays_act", "nplay_video", "nchapters", "nforum_posts"]

# 1) viewed == 1 인 경우
# 접근 이력은 있으나 활동량은 확정할 수 없으므로 활동지표는 건드리지 않음
# last_event_DI만 start_time_DI로 보정
course_fixed.loc[mask_viewed_missing, "last_event_DI"] = (
    course_fixed.loc[mask_viewed_missing, "start_time_DI"]
)

# 필요 시 플래그 추가
course_fixed["viewed_missing_flag"] = 0
course_fixed.loc[mask_viewed_missing, "viewed_missing_flag"] = 1

# 2) viewed != 1 인 경우
# 등록만 하고 추가 활동이 없었다고 보고 활동지표 0 처리
course_fixed.loc[mask_non_viewed_missing, activity_cols] = 0
course_fixed.loc[mask_non_viewed_missing, "last_event_DI"] = (
    course_fixed.loc[mask_non_viewed_missing, "start_time_DI"]
)

# 확인
print("Total missing last_event_DI:", mask_last_missing.sum())
print("Viewed=1 among missing:", mask_viewed_missing.sum())
print("Viewed!=1 among missing:", mask_non_viewed_missing.sum())

print(
    course_fixed.loc[
        mask_last_missing,
        ["viewed", "start_time_DI", "last_event_DI", "viewed_missing_flag"] + activity_cols
    ].head()
)
```

```python
# 3.3 duration 계산 및 음수 duration 제거
course_fixed["last_event_DI"] = pd.to_datetime(course_fixed["last_event_DI"])
course_fixed["start_time_DI"] = pd.to_datetime(course_fixed["start_time_DI"])
course_fixed["duration"] = (
    course_fixed["last_event_DI"] - course_fixed["start_time_DI"]
).dt.days

mask_negative_duration = course_fixed["duration"] < 0

print("duration < 0 개수:", mask_negative_duration.sum())

course_fixed = course_fixed.loc[~mask_negative_duration].copy()
```

```python
# 3.4 차후 분석에 이용될 수 있으니, fast_completion_flag로 마킹
mask_fast_completion = (
    (course_fixed["duration"] == 0) &
    (
        (course_fixed["explored"] == 1) |
        (course_fixed["certified"] == 1)
    )
)

course_fixed["fast_completion_flag"] = 0
course_fixed.loc[mask_fast_completion, "fast_completion_flag"] = 1

print("fast completion 수:", course_fixed["fast_completion_flag"].sum())
```

### `nchapters`, `nplay_video` 조건부 결측 처리

- `nchapters`와 `nplay_video`의 결측은 모두 같은 의미를 가진다고 보기 어렵다.
- 일부 결측은 실제 활동이 없었던 경우일 수 있지만,
  일부는 로그가 관측되지 않은 결과일 가능성도 존재한다.

- 따라서 남아 있는 결측값을 일괄적으로 0으로 대체하지 않고,
  **등록 후 추가 활동 없이 이탈한 것으로 해석 가능한 경우에만**
  조건부로 0 처리하였다.

처리 기준:
- `viewed != 1`
- `last_event_DI = start_time_DI`

해석:
- 해당 조건을 만족하는 사용자는 등록 이후 별도의 학습 행동이 관측되지 않은 집단으로 보고,
  `nchapters = 0`, `nplay_video = 0`으로 처리하였다.

- 반면, 강의 접근 이력이 있거나(`viewed = 1`) 추가 학습 가능성이 남아 있는 경우에는
  결측을 임의로 0으로 바꾸지 않고 유지하였다.

→ 결측값의 의미를 구분하여 보수적으로 처리하였다.

```python
# 복사본
course_conditional = course_fixed.copy()

# 조건:
# 1) viewed != 1
# 2) last_event_DI == start_time_DI
# → 등록 후 별도 활동 없이 이탈한 사용자로 해석 가능한 경우
mask_registered_only = (
    (course_conditional["viewed"] != 1) &
    (course_conditional["last_event_DI"] == course_conditional["start_time_DI"])
)

# 조건부 결측 처리 대상 컬럼
target_cols = ["nchapters", "nplay_video"]

# 해당 조건에서만 null을 0으로 대체
course_conditional.loc[mask_registered_only, target_cols] = (
    course_conditional.loc[mask_registered_only, target_cols].fillna(0)
)

# 확인
print("조건 적용 대상 수:", mask_registered_only.sum())
print()
print("처리 후 결측 개수:")
print(course_conditional[target_cols].isna().sum())
print()

print(
    course_conditional.loc[
        mask_registered_only,
        ["viewed", "start_time_DI", "last_event_DI", "nchapters", "nplay_video"]
    ].head()
)
```

### age 생성

연령 변수(age)는 다음과 같이 생성하였다.

YoB_DI(출생연도)와 start_time_DI(수강 시작일)의 연도를 이용
계산식:
→ age = start_year - YoB_DI

해당 값은 연 단위로 계산된 것으로, 정확한 만 나이가 아닌 분석용 근사 연령에 해당한다. 위에서 정의한 비현실적 값 처리 및 제한적 보정 과정을 반영한 최종 age를 분석에 활용하였다.

```python
# 복사본 생성

course_final = course_conditional.copy()

# start_year 생성
course_final["start_year"] = pd.to_datetime(
    course_final["start_time_DI"],
    errors="coerce"
).dt.year

# YoB 숫자형 변환
course_final["YoB"] = pd.to_numeric(course_final["YoB"], errors="coerce")

# age 1차 생성
# age = start_year - YoB
course_final["age_raw"] = course_final["start_year"] - course_final["YoB"]

# 원래부터 계산 불가능한 경우
course_final["age_missing_original"] = course_final["age_raw"].isna()
```

### YoB 기반 age 변수의 비현실적 값 처리

YoB_DI(출생연도)와 start_time_DI(수강 시작일)를 이용하여 학습자의 연령(age)을 계산한 결과, 0세, 1세 등 현실적으로 해석하기 어려운 값이 일부 존재하였다. 이러한 값은 단순한 극단값이라기보다 입력 오류 또는 비정상적인 기록에서 비롯된 것일 가능성이 높다.

특히, edX의 공식 이용약관에 따르면 서비스 이용자는 최소 13세 이상이어야 하며, 사용자는 해당 조건을 충족함을 보증해야 한다 (“You must be at least 13 years old to use the Service”, edX Terms of Service). 이러한 정책은 Children's Online Privacy Protection Act 등 아동 개인정보 보호 규제에 기반한 것으로, 온라인 플랫폼 전반에 공통적으로 적용되는 기준이다.

이에 따라 본 분석에서는 다음 기준을 적용하여 비현실적인 연령 값을 식별하였다.

13세 미만
음수 값
100세 초과

해당 조건에 해당하는 연령 값은 비현실적 값으로 간주하고, 우선 결측값(NaN)으로 처리하였다.

```python
# 비현실적 age 판정
course_final["age_invalid"] = (
    (course_final["age_raw"] < 13) |
    (course_final["age_raw"] < 0) |
    (course_final["age_raw"] > 100)
)


# 비현실적 age는 우선 NaN 처리
course_final["age_cleaned"] = course_final["age_raw"]
course_final.loc[course_final["age_invalid"], "age_cleaned"] = np.nan
```

### 비현실적 age에 대한 제한적 보정 (LoE 기반 대체)

비현실적 연령 값은 단순 제거할 경우 데이터 손실이 발생할 수 있으므로, 학위 수준(LoE_DI) 정보를 활용한 제한적 보정 방식을 적용하였다.

데이터 탐색 결과, 학위 수준이 높아질수록 연령이 증가하는 경향이 명확하게 나타났으며, 예를 들어 Bachelor's는 약 25세, Master's는 약 30세, Doctorate는 약 38세 수준의 중위 연령을 보였다. 이를 바탕으로 학위 수준은 연령을 추정하는 데 일정 수준 유효한 정보를 제공한다고 판단하였다.

따라서 다음과 같은 기준으로 보정을 수행하였다.

비현실적 연령으로 판정된 경우에 한함
LoE_DI 정보가 존재하는 경우에 한함
해당 학위 그룹의 중위 연령으로 대체

반면, YoB_DI 결측 등으로 인해 애초에 연령 계산이 불가능했던 경우는 별도의 추정 없이 결측 상태를 그대로 유지하였다. 이는 실제로 정보가 존재하지 않는 값까지 임의로 보정하는 것을 방지하기 위함이다.

### 한계 및 고려사항

해당 보정 방식은 데이터 손실을 줄이고 분석 활용도를 높이는 데 기여하지만, 몇 가지 한계를 가진다.

동일 학위 내에서도 다양한 연령 분포가 존재함에도 불구하고, 중위값으로 대체함에 따라 개별 학습자의 다양성이 일부 반영되지 못할 수 있다.
특정 연령 값으로 관측치가 집중되는 분포 왜곡이 발생할 수 있다.
학위와 연령 간의 관계가 일부 강화되는 효과가 나타날 수 있다.

그럼에도 불구하고 본 분석에서는, 해당 보정이 모든 결측값을 대상으로 한 일반적 대체가 아니라 비현실적 오류값에 한정된 제한적 보정이라는 점과, 연령 변수가 보조적인 설명 변수로 활용된다는 점을 고려하여 본 방식을 채택하였다.

```python
loe_median = (
    course_final.loc[course_final["age_cleaned"].notna()]
    .groupby("LoE_DI")["age_cleaned"]
    .median()
)


print("LoE_DI별 median age")
print(loe_median)

# 비현실적 age였던 값 중, LoE_DI가 있는 경우만 대체
course_final["age_final"] = course_final["age_cleaned"]

age_impute_mask = (
    course_final["age_invalid"] &
    course_final["LoE_DI"].notna()
)

course_final.loc[age_impute_mask, "age_final"] = (
    course_final.loc[age_impute_mask, "LoE_DI"].map(loe_median)
)

# 최종 분석용 age 컬럼
course_final["age"] = course_final["age_final"]
```

### age 구간화 변수 생성

연속형 변수인 age는 해석과 시각화의 편의를 위해 구간화하였다.

구간 기준은 다음과 같다.

20세 미만
20대
30대
40대
50대
60세 이상

이를 기반으로 범주형 변수 age_group을 생성하였으며, 연령 값이 존재하지 않는 경우는 "unknown"으로 분류하였다.

→ 연령대별 학습 패턴 및 성과 차이를 비교하기 위한 변수로 활용

```python
bins = [0, 19, 29, 39, 49, 59, 200]
labels = ["under_20", "20s", "30s", "40s", "50s", "60_plus"]

course_final["age_group"] = pd.cut(
    course_final["age_final"],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

course_final["age_group"] = (
    course_final["age_group"]
    .astype("object")
    .fillna("unknown")
)
```

### exam_flag 생성

- `grade` 값이 존재한다는 것은 최소한 평가 또는 시험 기록이 존재한다는 의미로 해석하였다.

처리 방식:

### LoE 기반 age 보정 방식에 대한 검증

비현실적인 연령 값에 대해 학위 수준(LoE_DI)을 기준으로 중위 연령을 활용한 보정 방식을 적용하기에 앞서, 해당 방법이 데이터 구조에 미치는 영향을 검토하였다. 이를 위해 연령 분포, 학위와 연령 간 관계, 분산 변화 등을 중심으로 검증을 수행하였다.

### LoE_DI와 age 간 관계 확인

먼저 학위 수준과 연령 간의 관계를 확인한 결과, 학위 수준이 높아질수록 연령이 증가하는 경향이 명확하게 나타났다. 예를 들어 Bachelor's, Master's, Doctorate 순으로 중위 연령이 점진적으로 증가하는 패턴이 관찰되었다.

이는 학위 정보가 연령을 추정하는 데 일정 수준의 설명력을 가진다는 것을 의미하며, LoE 기반 보정 방식의 적용 가능성을 뒷받침한다.

```python
loe_age = course_final.loc[
    course_final["LoE_DI"].notna() & course_final["age_cleaned"].notna(),
    ["LoE_DI", "age_cleaned"]
].copy()

loe_order = [
    "Less than Secondary",
    "Secondary",
    "Junior Secondary/Junior High/middle school",
    "Elementary/Primary school",
    "Bachelor's",
    "Master's",
    "Doctorate",
    "unknown"
]

loe_age["LoE_DI"] = pd.Categorical(
    loe_age["LoE_DI"],
    categories=loe_order,
    ordered=True
)

loe_age = loe_age.sort_values("LoE_DI")

print("\n학위별 age_cleaned 요약통계")
print(
    loe_age.groupby("LoE_DI", observed=False)["age_cleaned"]
    .agg(["count", "mean", "median", "std", "min", "max"])
    .sort_values("median")
)

plot_order = [
    loe for loe in loe_order
    if loe in loe_age["LoE_DI"].dropna().astype(str).unique()
]

data_to_plot = [
    loe_age.loc[loe_age["LoE_DI"] == loe, "age_cleaned"].dropna().values
    for loe in plot_order
]

plt.figure(figsize=(12, 6))
plt.boxplot(data_to_plot, labels=plot_order)
plt.xticks(rotation=45)
plt.ylabel("Age (cleaned)")
plt.title("Age_cleaned by LoE_DI")
plt.grid(False)
plt.show()
```

### 분산(variance) 변화 분석
보정 전후의 연령 분산을 비교한 결과, 분산 값은 약 77.94에서 77.89로 거의 변화가 없는 수준으로 나타났다.

이는 일부 관측치에 대해 학위별 중위값으로 대체가 이루어졌음에도 불구하고, 전체 데이터의 변동성에는 유의미한 영향을 미치지 않았음을 의미한다. 즉, 보정 과정이 특정 값으로의 과도한 집중을 유발하거나 분포를 인위적으로 축소시키는 수준은 아닌 것으로 판단된다.

### 학위와 연령 간 상관관계 확인
학위 수준과 연령 간의 상관계수를 비교한 결과, 보정 전후 값이 각각 약 0.499 수준으로 거의 동일하게 유지되는 것으로 나타났다.

이는 기존 데이터에서도 학위와 연령 간에 일정한 양의 관계가 존재했음을 의미하며, 보정 과정이 새로운 관계를 인위적으로 강화하거나 왜곡하지 않았음을 시사한다. 따라서 LoE 기반 보정은 기존 데이터 구조를 유지하는 범위 내에서 이루어진 것으로 해석할 수 있다.

```python
# 검증 분산 비교
print("\nVariance 비교")
print("age_cleaned variance:", course_final["age_cleaned"].var())
print("age_final variance:", course_final["age_final"].var())

# 검증 LoE와 age 상관 비교
loe_map = {
    "Less than Secondary": 1,
    "Secondary": 2,
    "Bachelor's": 3,
    "Master's": 4,
    "Doctorate": 5
}

course_final["LoE_num"] = course_final["LoE_DI"].map(loe_map)

cleaned_corr = course_final[["LoE_num", "age_cleaned"]].corr().iloc[0, 1]
final_corr = course_final[["LoE_num", "age_final"]].corr().iloc[0, 1]

print("\nCorrelation 비교")
print("LoE_num vs age_cleaned:", cleaned_corr)
print("LoE_num vs age_final:", final_corr)
```

## 파생 변수 정의 (Derived Features)

본 전처리 과정에서는 기존 변수들을 기반으로
분석 및 해석을 용이하게 하기 위한 파생 변수들을 생성하였다.

---

### 1. age

- 정의: 학습자의 연령 (추정값)

- 생성 방식:
  - `age = start_year - YoB_DI`

- 설명:
  - `YoB_DI`(출생연도)와 `start_time_DI`(수강 시작일)를 기반으로 계산
  - 정확한 생일 정보가 없기 때문에 **만 나이가 아닌 연 단위 근사값**

- 처리 보완:
  - 비현실적인 값 (`age < 0`, `age > 100`)은 결측 처리

→ 학습자 연령 분포 및 연령별 행동 분석에 활용

---

### 2. age_group

- 정의: 연령 구간 범주 변수

- 생성 방식:
  - `age`를 다음 구간으로 분할:
    - under_20
    - 20s
    - 30s
    - 40s
    - 50s
    - 60_plus

- 설명:
  - 연속형 변수인 `age`를 범주형으로 변환하여
    해석 및 시각화 용이성 확보

- 추가 처리:
  - 결측값은 `"unknown"`으로 대체

→ 연령대별 학습 행동 및 이탈 패턴 비교 분석에 활용

---

### 3. exam_flag

- 정의: 평가(시험) 참여 여부

- 생성 방식:
  - `grade`가 존재하면 1
  - `grade`가 결측이면 0

- 설명:
  - `grade` 값은 단순 점수뿐 아니라
    **평가 또는 시험 참여 여부를 나타내는 신호**로 해석 가능

→ 학습 참여도 및 성취도 분석을 위한 보조 지표

---

### 4. duration

- 정의: 학습 기간 (일 단위)

- 생성 방식:
  - `duration = last_event_DI - start_time_DI`

- 설명:
  - 수강 시작일부터 마지막 활동일까지의 기간
  - 학습 지속 기간을 나타내는 핵심 변수

- 처리 보완:
  - `duration < 0`인 경우는 시간 역전 데이터로 간주하여 제거

→ 학습 지속성 및 이탈 시점 분석에 활용

---

### 5. viewed_missing_flag

- 정의: viewed는 존재하지만 활동 로그가 관측되지 않은 사용자 식별 변수

- 생성 방식:
  - `viewed = 1` 이면서 `last_event_DI`가 결측인 경우 → 1
  - 그 외 → 0

- 설명:
  - 최소 1회 접근은 있었으나
    활동 로그(`nevents`, `ndays_act` 등)가 관측되지 않은 경우

- 해석:
  - 데이터 수집 한계 또는 부분 관측 데이터

→ 분석 시 별도 그룹으로 분리하여 해석 가능

---

### 6. fast_completion_flag

- 정의: 단기 완료 사용자 식별 변수

- 생성 방식:
  - `duration = 0` 이면서
    `explored = 1` 또는 `certified = 1`인 경우 → 1

- 설명:
  - 하루 내 집중 학습 또는 빠른 수료 패턴을 나타냄

- 처리 결정:
  - 비정상 데이터로 제거하지 않고
    **특이한 학습 패턴으로 간주하여 유지**

→ 고속 학습자(heavy learner) 분석에 활용 가능

---

## 범주형 결측값 처리

다음 변수들은 결측을 제거하지 않고 `"unknown"`으로 대체하였다:

- `LoE_DI` (교육 수준)
- `gender`
- `age_group`

설명:
- 해당 결측은 단순 오류가 아닌 정보 부재일 가능성이 높음
- 행 제거 시 데이터 손실이 발생하므로 별도 범주로 유지

→ 결측을 하나의 의미 있는 집단으로 포함하여 분석 수행
