```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
```

```python
current_path = os.getcwd()
parent_path = os.path.dirname(current_path)
data_path = os.path.join(parent_path, 'data')

portfolio = pd.read_csv(data_path + '\\Portfolio_clean.csv')
profile = pd.read_csv(data_path + '\\profile_clean.csv')
transcript = pd.read_csv(data_path + '\\transcript_clean.csv')
```
